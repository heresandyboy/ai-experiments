```python
import requests
import json
import re

def generate_response_http(query, top_p):
    # endpoint = os.environ.get("ENDPOINT", None)
    # token = os.environ.get("TOKEN", None)
    endpoint = "https://api-inference.huggingface.co/models/bigcode/starcoder/"
    params = {
        'accept': 'application/json',
        'Content-Type': 'application/json',
        "Authorization": f"Bearer {HF_HUB_TOKEN}"
    }
    data = {
        'inputs': query,
        "parameters": {
            "details": True,
            "do_sample": True,
            "max_new_tokens": 256,
            "repetition_penalty": 1.2,
            "seed": 0,
            "stop": ['<|endoftext|>'],
            "temperature": 0.1,
            "top_p": top_p,
        }
    }
    try:
        res = requests.post(endpoint, data=json.dumps(data), headers=params)
        res = json.loads(res.content.decode("utf-8"))
        return res[0]["generated_text"].split("<|endoftext|>")[0].strip()
    except:
        raise print("Connection Error. Please check your ACCESS_TOKEN")
```

```python
def generate_response(input_prompt,top_p):
    inputs = tokenizer.encode(input_prompt, return_tensors="pt").to('cuda')
    outputs = model.generate(inputs,
                            eos_token_id = 0,
                            pad_token_id = 0,
                            max_new_tokens=256,
                             top_p= 0.9,
                             temperature =0.2,
                             repetition_penalty = 1.2,
                            early_stopping=True)
    output =  tokenizer.decode(outputs[0])
    return output
```

```python
def post_process_code(code, question):
    parameters = code.split("\n")[0].split("def solution")[-1][1:-2].split(",")
    if '' in parameters:
        parameters.remove('')
    values = re.findall(r"[-+]?\d*\.\d+|\d+", question)[:len(parameters)]
    values = [int(v) for v in values]
    return list(zip(parameters, values))


def solve_pal(question):
    question = question.strip()
    print(question)
    query = long_prompt.format(question=question).strip()
    code = generate_response(query, 0.9)
    # code = generate_response_http(query, 0.9)
    # print(code)
    code = code.split("def solution():")[-1].strip()
    code = "def solution():\n" + code
    code = code.replace("<|endoftext|>", '')
    code = code.replace("<jupyter_output><empty_output>",'')
    # code preprocessing

    arguments = post_process_code(code, question)
    # print(arguments)
    

    arg_string = ""
    for param, val in arguments:
        arg_string += f"{param}={val},"
    func_call = f"\nprint(solution({arg_string[:-1]}))"
    code += func_call
    # code running
    if "input(" in code:
        return None, code
    pred = None
    try:
        pred = run_code(code)
    except Exception as ex:
        return None, code
    return pred, code
```

```python
from typing import Dict, Union, Any, List

from langchain.chains import PALChain
from langchain import PromptTemplate, LLMChain
from langchain.callbacks.base import BaseCallbackHandler


class MyCustomHandler(BaseCallbackHandler):
    def on_llm_start(
        self, serialized: Dict[str, Any], prompts: List[str], **kwargs: Any
    ) -> Any:
        """Run when LLM starts running."""
        print(f"My custom handler, llm_start: {prompts[-1]} stop")

    def on_llm_end(self, response: Any, **kwargs: Any) -> Any:
        """Run when LLM ends running."""
        print(f"My custom handler, llm_end: {response.generations[0][0].text} stop")

```
